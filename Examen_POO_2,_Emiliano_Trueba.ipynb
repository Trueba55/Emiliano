¿Qué es una clase y por qué es útil? \\

**Respuesta Ejercicio 1**
:
Una clase es como un manual que te dice como debe ser algo y que puede hacer un objeto, un objeto son instancias de una clase, y cada objeto creado a partir de una clase hereda sus atributos y métodos, lo que permite organizar el código de manera eficiente.


```python
#Respuesta Ejericio 1
class coche:
    def __init__(self,marca, modelo, año):
        self.marca = marca
        self.modelo = modelo
        self.año = año

    def obtener_informacion(self):
        return f"Coche {self.marca}, modelo {self.modelo} del año {self.año}"

# Crear un objeto de la clase coche
mi_coche = coche("Mazda","Mazda 3 HB", 2023)

# Obtener información sobre el coche
print(mi_coche.obtener_informacion())
```
